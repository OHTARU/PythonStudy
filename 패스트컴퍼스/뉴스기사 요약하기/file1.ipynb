```python
import base64
```

```python
string = "Life is too short. We need Python"
```

```python
encoded = base64.b64encode(string)
```

```python
bstring = string.encode('ascii')
print(bstring)
```

```python
encoded = base64.b64encode(bstring)
```

```python
print(encoded)
```

```python
decoded = base64.b64decode(encoded)
```

```python
print(decoded)
```

```python
decoded_str = decoded.decode('ascii')
print(decoded_str)
```
